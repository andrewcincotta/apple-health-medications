```python
# Import packages
import pandas as pd
import numpy as np
import json
```

```python
# Import CSV Provided in pandas
inputFile = "csv/Medications-2025-04-23-2026-04-23.csv"
outputFile = "csv/cleaned/Meds-2025-04-23-2026-04-23.csv"
df = pd.read_csv(inputFile)
```

```python
# print cols
df.columns
# print medication names
df['Medication'].unique()
# print rows where status is "test"
df[df['Status'] == 'test']
```

```python
# delete bad cols such as Scheduled Date
# if scheduled date doesnt equal "Taken" remove row from df
if df['Scheduled Date'][0] != 'Taken':
    df.drop(df.index[0], inplace=True)

df.drop(['Scheduled Date'], axis=1, inplace=True)
df.drop(['Nickname'], axis=1, inplace=True)
df.drop(['Scheduled Dosage'], axis=1, inplace=True)
df.drop(['Status'], axis=1, inplace=True)
df.drop(['Archived'], axis=1, inplace=True)
df.drop(['Codings'], axis=1, inplace=True)
df.drop(['Unit'], axis=1, inplace=True)

# Rename dosage col to Count
df.rename(columns={'Dosage': 'Count'}, inplace=True)

# Print new cols
df.columns
```

```python
# Load mappings from JSON
with open('../config/default_medication_map.json', 'r') as f:
    mappings = json.load(f)

MedsToNicknames = mappings.get('MedsToNicknames', {})
NicknamesToDosage = mappings.get('NicknamesToDosage', {})
NicknameToCost = mappings.get('NicknameToCost', {})
```

```python
# Create Nickname and "Unit (mg)" cols using mapping based on previous cols
df['Nickname'] = df['Medication'].map(MedsToNicknames)
df['Unit (mg)'] = df['Nickname'].map(NicknamesToDosage)

# print cols again
df.columns
```

```python
# Create calculated column of "Dosage (mg)" based on Count and "Unit (mg)
df['Dosage (mg)'] = round(df['Count'] * df['Unit (mg)'], 3)

# print cols again
df.columns
```
