```python
import pandas as pd
import numpy as np
```

```python
sheet = "ExcelReport.csv"
df = pd.read_csv(sheet)
df.columns
```

```python
# Delete Type Scheduled for and notes columns
df = df.drop(columns=['Type', 'Scheduled for', 'Notes'])

# Rename Recorded on to Date
df = df.rename(columns={'Recorded on': 'Date'})

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

df.columns
```

```python
# delete rows where Value != Taken
prevLen = len(df)
df = df[df['Value'] == 'Taken']
print(f"Deleted {prevLen - len(df)} rows")
```

```python
# Create new column, Count, with values of 1
df['Count'] = 1
```

```python
# Create Medication Mapping of count to mg
df['Name'].unique()

countToMG = {
    "Cannabis": 1,
    "Klonopin": 0.5,
    "Phenibut": 500,
    "Xanax": 0.5,
    "Zenzedi": 5
}

# Ensure 'Count' column exists and has values of 1
df['Count'] = 1

# Create Dose column as the product of Count and Name's mapped mg
df['Dose (mg)'] = df['Count'] * df['Name'].map(countToMG)
```

```python
# Delete Value row
df.drop(columns=['Value'], inplace=True)

# Rename Cannabis to Weed and Zenzedi to Dexedrine
df.replace({'Cannabis': 'Weed', 'Zenzedi': 'Dexedrine'}, inplace=True)
```

```python
# Set column order to Name, Date, Count, Dose (mg)
df = df[['Name', 'Date', 'Count', 'Dose (mg)']]

# Sort by Date high to low
df = df.sort_values(by='Date', ascending=False)

# drop last entry of cannabis
df.drop(0, inplace=True) # drop latest weed log

# Reset index
df.reset_index(drop=True, inplace=True)
df.head(10)
```

```python
# Save transformed df as a csv
df.to_csv('transformed.csv', index=False)
```
